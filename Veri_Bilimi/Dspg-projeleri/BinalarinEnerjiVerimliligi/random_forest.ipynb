```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import *
```

```python
data=pd.read_csv("data.csv")
```

```python
Y1=data["Y1"]
Y2=data["Y2"]
X=data.iloc[:,0:8]
```

```python
model1=RandomForestRegressor()
model2=RandomForestRegressor(max_features="sqrt")
for _ in range(150):
    X_train,X_test,Y_train1,Y_test1=train_test_split(X,Y1,train_size=0.80)
    Y_test2=Y2[Y_test1.index]
    Y_train2=Y2[Y_train1.index]
    
    model1.fit(X_train.values,Y_train1.values)
    model2.fit(X_train.values,Y_train2.values)
    
    y1_predict=model1.predict(X_test.values)
    y2_predict=model2.predict(X_test.values)
    y1_error = mean_absolute_error(y1_predict,Y_test1.values)
    y2_error = mean_absolute_error(y2_predict,Y_test2.values)
    if y1_error < 0.30 and y2_error <0.82:
        break
    print({"Heating Load":y1_error,"Cooling Load":y2_error})
```

```python
class Model:
    def __init__(self,model1,model2):
        self.model1=model1
        self.model2=model2
    def predict(self,X):
        return {"Heating Load":self.model1.predict(X)[0],"Cooling Load":self.model2.predict(X)[0]}
```

```python
model = Model(model1,model2)
model.predict(X_test.iloc[0].values.reshape(1,-1))
```
