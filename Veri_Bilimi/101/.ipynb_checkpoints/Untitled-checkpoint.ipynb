```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("data/Train.csv",parse_dates=["saledate"],low_memory=False)
```

```python
df
```

```python
df["UsageBand"].value_counts()
```

```python
def train_cats(df):
    for n,c in df.items():
        if c.dtype=="object":
            df[n]=c.astype("category").cat.as_ordered()
```

```python
train_cats(df)
```
