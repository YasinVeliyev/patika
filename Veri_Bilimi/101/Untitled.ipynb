```python
import pandas as pd
import numpy as np
from pandas.api.types import is_string_dtype,is_numeric_dtype
```

```python
df=pd.read_csv("data/Train.csv",parse_dates=["saledate"],low_memory=False)
```

```python
def train_cats(df):
    for n,c in df.items():
        print(n)
        if is_string_dtype(c):
            df[n]=c.astype("category").cat.as_ordered()
```

```python
def apply_cats(df,train):
    for n,c in df.items():
        print(n)
        if train[n].dtype=="category":
            df[n]=pd.Categorical(c,categories=train[n].cat.categories,ordered=True)
```

```python
def numericalize(df,col_name):
    if df[col_name].dtype=="category":
        df[col_name]=df[col_name].cat.codes+1
```

```python
def add_datepart(df,col_name,drop=True):
    dt_column=df[col_name]
    col_dtype=dt_column.dtype
    attr=["year","month","week","day","dayofweek","dayofyear",'is_month_end','is_month_start','is_quarter_end','is_quarter_start','is_year_end','is_year_start']
    for a in attr:
        df["Date" + a.capitalize()]=getattr(dt_column.dt,a)
    df["Date"+"Elapsed"] = dt_column.astype(np.int64)//10**9
    if drop:
        df.drop(col_name,inplace=True,axis=1)
```

```python
df["UsageBand"].cat.set_categories(["High","Medium","Low"],ordered=True,inplace=True)
```

```python
def fix_missing_value(df,col_name,nan_dict,is_train):
    col = df[col_name]
    if is_train:
        if is_numeric_dtype(col):
            if col.isnull().sum():
                df[col_name+"_na"]=col.isnull()
                nan_dict[col_name]=col.median()
                df[col_name]=col.fillna(nan_dict[col_name])
    else:
        if is_numeric_dtype(col):
            if col_name in nan_dict:
                df[f"{col_name}_na"]=col.isnull()
                df[col_name]=df[col_name].fillna(nan_dict[col_name])
            else:
                df[col_name]=col.fillna(col.median())
```

```python
def proc_df(df,y_fld,nan_dict=None,is_train=True):
    df=df.copy()
    y=df[y_fld].values
    df.drop([y_fld],axis=1,inplace=True)
    if not nan_dict:
        nan_dict={}
        
    for n,c in df.items():
        fix_missing_value(df,n,nan_dict,is_train)
        numericalize(df,n)
    if is_train:
        return df,y,nan_dict
    return df,y
```

```python
def split_train_val(df,n):
    return df[:n].copy(),df[n:].copy()
```

```python
n_valid=12000
n_train=len(df)-n_valid
raw_train,raw_valid=split_train_val(df,n_train)
x_train,y_train,nas=proc_df(raw_train,"SalePrice")
```
